```python
# We can notice that in almost all variables the peak point is around 2000 but in volume it is not because in the raw data we have,
# the values of feature volume are way too big in comparison to other features.
# So, let's normalize the data using simple feature scaling. It will also be helpful in ML model building later.
```

```python
df['Open'] = df['Open']/df['Open'].max()
df['High'] = df['High']/df['High'].max()
df['Close'] = df['Close']/df['Close'].max()
df['Low'] = df['Low']/df['Low'].max()
df['Adj Close'] = df['Adj Close']/df['Adj Close'].max()
df['Volume'] = df['Volume']/df['Volume'].max()
```
